# **5. Baseline regression: predict temperature**
#### **Goal: predict Temperature (C) from a single timestampâ€™s features (no lag yet)**

#### **5.1 Prepare data for Regression**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
target = "Temperature (C)"
feature_cols = [
    "Year", "Month", "Day", "Hour", "DayOfWeek",
    "Humidity", "Wind Speed (km/h)", "Wind Bearing (degrees)",
    "Visibility (km)", "Pressure (millibars)",
    "Precip Type", "Summary"  # Summary will be simplified
]
```

```python
df_model = df[feature_cols + [target]].copy()
```

```python
df_model.head()
```

#### **Simple Text Cleanup**

```python
#reduce Summary column to a few categories
def simplify_summary(s):
    s = s.lower()
    if "rain" in s: return "rain"
    if "snow" in s: return "snow"
    if "fog" in s: return "fog"
    if "clear" in s: return "clear"
    if "cloud" in s: return "cloudy"
    return "other"

df_model["Summary"] = df_model["Summary"].astype(str).apply(simplify_summary)
```

```python
df_model.head()
```

#### **Split Features and Target**
where:
* X contains all predictor columns (features_cols).
* y is the target variable you want to predict.



```python
X = df_model.drop(columns=[target])
y = df_model[target]
```

```python
df_model.shape
```

### **Train-Validate-test Split**

#### **1. Split dataset into training + remaining (validation + test)**
* X_train, y_train: 70% of data for training.
* X_rem, y_rem: 30% of data that will be split into validation and test
* shuffle=False: preserves temporal order for time series tasks.

```python
# First split: train vs remaining
X_train, X_rem, y_train, y_rem = train_test_split(
    X, y, test_size=0.3, shuffle=False  # keep time order if time series
)
```

#### **2. Split remaining into validation and test**
* X_val, y_val: 15% of original data for cross-validation.
* X_test, y_test: 15% of original data for final testing.
* test_size=0.5 splits the 30% remaining evenly.

```python
# Second split: validation vs test
X_val, X_test, y_val, y_test = train_test_split(
    X_rem, y_rem, test_size=0.5, shuffle=False
)
```

```python
#check shape of the splitted features
X_test.shape, X_rem.shape, X_train.shape, X_val.shape
```

```python
df_model.shape
```

#### **Identify Categorical and Numerical Features**
where:
* cat_cols lists features that are categorical and need encoding.
* num_cols contains all remaining numeric columns


```python
cat_cols = ["Precip Type", "Summary", "Month", "DayOfWeek", "Hour"]
num_cols = [c for c in X.columns if c not in cat_cols]
```

#### **Preprocessing pipeline (categorical)**
* Categorical columns are one-hot encoded.
* To avoids errors if validation/test sets contain new categories, set handle_unknown="ignore"


```python
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ]
)
```

#### **5.2 Baseline model: Linear Regression and Random Forest**

##### **Evaluation function**
* Trains Model only on training data
* Computed Metrics on validation data
* No test data touched yet

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def eval_reg_model(model, X_train, y_train, X_val, y_val, name="model"):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    mae = mean_absolute_error(y_val, y_pred)
    rmse = mean_squared_error(y_val, y_pred, squared=False)

    print(f"{name} | Validation MAE: {mae:.3f}, RMSE: {rmse:.3f}")
```

##### **Build Pipelines**
Why pipelines matter
* Preprocessing is learned only from training data
* Encoding stays consistent across splits
* No data leakage

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

lin_pipeline = Pipeline([
    ("preprocess", preprocess),
    ("model", LinearRegression())
])

rf_pipeline = Pipeline([
    ("preprocess", preprocess),
    ("model", RandomForestRegressor(
        n_estimators=200,
        random_state=0,
        n_jobs=-1
    ))
])
```

##### **Model Evaluation**
This idea here is:
* Evaluates both models on the same validation set for fair comparison.
* Helps decide which model generalizes better before using the test set.
* Separates model selection (validation) from final assessment (test).

```python
eval_reg_model(lin_pipeline, X_train, y_train, X_val, y_val, "Linear Regression")
eval_reg_model(rf_pipeline, X_train, y_train, X_val, y_val, "Random Forest")
```
